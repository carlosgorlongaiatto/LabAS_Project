# Naive Bayes classifier for fake news recognition

The goal of the project is to implement a Multinomial Naive Bayes classifier in R and test its performances in the classification of social media posts.

### Dataset

We analyze a [dataset](https://www.kaggle.com/datasets/anmolkumar/fake-news-content-detection?select=train.csv) available on Kaggle that consists of more than 10000 text documents representing content from social media and labeled as follows:

0 -> Barely-True \
1 -> False \
2 -> Half-True \
3 -> Mostly-True \
4 -> Not-Known \
5 -> True

```r
# set seed
seed <- 54321
set.seed(seed)

# import data
data <- read.csv('data/fake_news.csv')

# select a subset of the data
#data <- data[sample(nrow(data), 5000), ]

# merge close labels
#data$Labels <- sapply(data$Labels, function(x) switch(x+1, 0, 0, 1, 2, 1, 2))

# get labels
label_list <- sort(unique(data$Labels))

# print some info
cat('number of rows:', nrow(data), '\n')
cat('number of columns:', ncol(data), '\n')
head(data, 5)
```

### Training, validation and test sets

The dataset is divided into training set, validation set and testing set.
The training set is used to learn the model from the data and the validation set provides unbiased evaluation while tuning the hyperparameters of the model. Finally, the test set is used to evaluate the performance of the best model.

```r
# split training, validation and test set
index <- sample(seq(1, 3), size = nrow(data), replace = TRUE, prob = c(0.5, 0.3, 0.2))

train <- data[index == 1,]
val <- data[index == 2,]
test <- data[index == 3,]
```

```r
# plot the percentage frequency of each label for each set
plot_label_freq(list(training = train, validation = val, test = test))
```

### Tokenization

Each document in our dataset is a string of characters that must be preprocessed before being fed into the algorithm. In particular we convert the uppercase letters of a string into lowercase letters and then chop it up into chunks, called tokens, throwing away anything that is not a letter. This represents the first step in the preprocessing stage, which, however, is not without its problems. For example, chopping on apostrophe for possession and contractions could lead to incorrect tokens. These types of errors will be handled later. Here is an example of tokenization:

```r
document <- train$Text[20]
document_step1 <- tokenize(document)

cat('document:', '\n')
cat(document, '\n')

cat('\ntokenized document:', '\n')
print(document_step1)
```

### Stop words

The second step is to clean up any collection of tokens from stop words. The stop words are some extremely common words that seem to be of little value in helping to properly classify documents.  We also remove the single letters produced during the tokenization. Here is an example:

```r
document_step2 <- cleaning(document_step1)

cat('tokenized document:', '\n')
print(document_step1)

cat('\ntokenized document without stop words:', '\n')
print(document_step2)
```

### Lemmatization

The last step in the preprocessing phase is what is known as token normalization, which consists of creating equivalence classes so that similar tokens are mapped to the same class. In particular, we use the lemmatization method, which performs a complete morphological analysis to accurately identify the lemma of each word, such as removing inflectional endings. Here is an example:


```r
document_step3 <- lemmatize_words(document_step2)

cat('tokenized document without stop words:', '\n')
print(document_step2)

cat('\ntokenized document without stop words and after lemmatization:', '\n')
print(document_step3)
```

### Vocabulary

After the pre-processing stage, we need to build the so-called vocabulary, which is the unique collection of all tokens that appear in the training set and will be used for classification.

```r
# pre-process documents in training set
train$Tokens <- v_pre_processing(train$Text, CLEAN = TRUE, LEM = TRUE)

# make vocabulary
vocabulary <- make_vocabulary(train)
```

### Feature selection: mutual information

The last stage before delving into the algorithm is represented by feature selection. Feature selection is the process of selecting a subset of the tokens in the vocabulary extracted from the training set and using only this subset as features in the classification of new documents. It is used to reduce the high-dimensional document space and eliminate noise features. This prevents the algorithm from overfitting and provides better generalization in the test set. In particular we compute the utility measure as the expected mutual information (MI) of each token and each label. MI measures how much information presence/absence of a term contributes to making the correct classification decision on a selected label. The formula is the following one:


$$ \text{MI}(X,Y) = \sum_{x \in X} \sum_{y \in Y} P_{X,Y}(x,y) \cdot \log_2\bigg(\frac{P_{X,Y}(x,y)}{P_{X}(x) \cdot P_{Y}(y)}\bigg) $$

where $X$ and $Y$ are discrete variables, $P_{X,Y}(x,y)$ is the joint distribution and $P_{X}(x)$ and $P_{Y}(y)$ are the marginal distributions. Then we are left with a vocabulary for each label sorted by mutual information. The idea is to select the k terms with the largest values of MI for each label and then update the vocabulary of the entire training set by again taking a unique collection of all tokens.

```r
# mutual information
#mutual_information <- get_mutual_information(train, vocabulary)
#write.csv(mutual_information, 'objects/mutual_information_1.csv')

# read mutual information from file
mutual_information <- as.list(read.csv('objects/mutual_information_1.csv', row.names = 1))
for (i in 1:length(mutual_information)) {names(mutual_information[[i]]) <- vocabulary}
```

So far we have considered two stages: the pre-processing and the feature selection. To decide whether it is better to apply the cleaning and lemmatization procedures and to set the optimal value of the hyperparameter k, we perform a grid search on the validation set. It is now necessary to describe the algorithm we developed, how we modeled it on the training set, and how we evaluated its performance on the test set.

### Multinomial naive Bayes model

In the multinomial naive Bayes model a document is assigned to a label by evaluating the maximum a posteriori probability, which is computed as follows using the **Bayes theorem**:



$$ \hat{l} = \arg \max_{l \in \mathcal{L}} P(l|d)
        = \arg \max_{l \in \mathcal{L}} \frac{P(d|l) \cdot P(l)}{P(d)}
        = \arg \max_{l \in \mathcal{L}} P(d|l) \cdot P(l) $$


where $d$ denotes a document and $l$ a label. Note that in the last step we remove the evidence factor in the denominator because it is the same for all labels and therefore argmax does not depend on it.

The next step is to make the **conditional independence assumption**, which is to assume that each token within a document is independent of each other for a fixed label. This is not actually true for text data, but the performance of the naive bayes classifier can be remarkably good despite this very strong assumption. Thus the likelihood becomes:

$$ P(d|l) = P(<t_1, \dots, t_{n_d}>|l) = \prod_{1<k<n_d} P(t_k|l) $$

where $n_d$ is the number of tokens in he document $d$. Another assumption we make is the **positional independence assumption**, which means that the conditional probability of a token is independent of its position in the document. This model is commonly known as the bag-of-words model.

After the above two assumptions, we can estimate both the priors and the conditional probabilities respectively as:

$$ P(l) = \frac{N_l}{N} \hspace{1mm} \forall \hspace{1mm} l \in \mathcal{L} $$

$$ P(t|l) = \frac{T_{l,t}}{\sum_{t' \in V} T_{l,t'}}
\hspace{1mm} \forall \hspace{1mm} l \in \mathcal{L} 
\hspace{1mm} \hspace{1mm} \text{and} \hspace{1mm} \forall \hspace{1mm} t \in \text{V} $$

where $T_{l,t}$ is the number of occurrences of $t$ in training documents from label $l$, including multiple occurrences of a token in a document, and $V$ is the vocabulary. It may happen that a token from the vocabulary does not occur in the documents of the training set for a given label. This implies that the maximum likelihood estimate related to that label for a document containing that specific token will always be zero, an effect that goes by the name of sparseness. In order to eliminate zeros, we use **add-one** or **Laplace smoothing**, which is as follows:

$$ P(t|l) = \frac{T_{l,t} + 1}{\sum_{t' \in V} (T_{l,t'}+1)}
        = \frac{T_{l,t} + 1}{\sum_{t' \in V} T_{l,t'} + |V|} $$

Finally, we can construct a matrix containing all conditional probabilities for each given label and for all tokens in the vocabulary.

Once we have trained our model, applying it to the test set is straightforward. Each new document is preprocessed and then posterior probabilities are estimated for each possible label using the priors and conditional probabilities calculated in the training phase. In the end, the winning label is assigned to the document. Note that if there is a token in a document in the test set that was not seen during the training phase, that is, it is not in the vocabulary, it is handled using Laplace smoothing again.

To evaluate the performance of the mutinomial naive Bayes classifier we use accuracy, defined as the ratio of correctly classified documents to the total number of documents. we also calculate the micro-averaging F1-measure, but since this is a one-of calssification, that is, the labels are assumed to be mutually exclusive, it is equal to the accuracy. Finally, we evaluate the confusion matrix to see whether the algorithm is confusing any label pairs.

### Grid search

```r
# pre-processing (cleaning and lemmatization) grid
pp_grid <- list(c(FALSE,FALSE), c(TRUE,TRUE))

# k grid
k_grid <- seq(0.1, 1, 0.1)

# grid search
#evaluation_grid <- grid_search(train, val, pp_grid, k_grid)
```

```r
# save evaluation grid
#colnames(evaluation_grid) <- k_grid
#write.csv(evaluation_grid, 'objects/evaluation_grid.csv', row.names=F)

# read evaluation grid from file
evaluation_grid <- as.matrix(read.csv('objects/evaluation_grid.csv'))
```

```r
best <- which(evaluation_grid == max(evaluation_grid), arr.ind = TRUE)

# select the best pre-processing
best_pp <- pp_grid[[best[1,1]]]
best_CLEAN <- best_pp[1]
best_LEM <- best_pp[2]

# select the best k
best_k <- k_grid[best[1,2]]
```

```r
if (best_CLEAN) { cat('cleaning ON\n') } else { cat('cleaning OFF\n') }
if (best_LEM) { cat('lemmatization ON\n') } else { cat('lemmatization OFF\n') }
cat('best k:', best_k)
```

```r
# plot grid search
plot_grid_search(pp_grid, k_grid, evaluation_grid)
```

### Results

```r
# pre-process documents in training set
train$Tokens <- v_pre_processing(train$Text, CLEAN = best_CLEAN, LEM = best_LEM)
```

```r
# make vocabulary
vocabulary <- make_vocabulary(train)
```

```r
# mutual information
#mutual_information <- get_mutual_information(train, vocabulary)
#write.csv(mutual_information, 'objects/mutual_information_2.csv')

# read mutual information from file
mutual_information <- as.list(read.csv('objects/mutual_information_2.csv', row.names = 1))
for (i in 1:length(mutual_information)) {names(mutual_information[[i]]) <- vocabulary}
```

```r
# features selection
k <- best_k
res <- features_selection(mutual_information, k)

selected_features <- res$selected_features 
vocabulary <- res$vocabulary
```

```r
# visualize top 10 features for each label
label_names <- c('Barely_True', 'False', 'Half_True', 'Mostly_True', 'Not_Known', 'True')
feature_ranking <- get_feature_ranking(selected_features, label_names)
head(feature_ranking, 10)
```

```r

load(file = 'functions')
plot_MI(mutual_information, k, 1500)
```

```r
# learning model
res <- learning_model(train, vocabulary)

conditional_probs <- res$conditional_probs
priors <- res$priors
tot_tokens_per_label <- res$tot_tokens_per_label
```

```r
# predict
test_set <- test
true_labels <- test_set$Labels
predicted_labels <- v_predict(test_set$Text, 
                    vocabulary = vocabulary,
                    conditional_probs = conditional_probs, priors = priors,
                    tot_tokens_per_label = tot_tokens_per_label,
                    CLEAN = best_CLEAN, LEM = best_LEM)
```

```r
# evaluate
res <- evaluate(true_labels, predicted_labels)

accuracy <- res$accuracy
confusion_matrix <- res$confusion_matrix

cat('accuracy:', accuracy)
plot_confusion_matrix(confusion_matrix)
```
