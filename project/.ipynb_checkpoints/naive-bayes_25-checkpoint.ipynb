# Functions

```r
plot_label_freq <- function(dataset) {

    # plot ordered mutual information for each class
    options(repr.plot.width = 15, repr.plot.height = 5)
    par(mfrow = c(1, 3))

    col <- brewer.pal(6, 'Dark2')

    for (i in 1:3) {

        label_freq <- dataset[[i]] %>% 
            group_by(Labels) %>%
            summarise(counts = length(Labels))

        barplot(label_freq[[2]]/sum(label_freq[[2]]), names.arg = label_freq[[1]], col = col,
            main = names(dataset)[i], xlab = 'label', ylab = 'frequency [%]',
            cex.lab = 1.5, cex.axis = 1.5, cex.names = 1.5, cex.main = 2)

    }
}
```

```r
# define function to tokenize document (bag of words model)
tokenize <- function(document) {

    # lowercase
    temp <- tolower(document)

    # remove everything that is not a letter
    temp <- str_replace_all(temp, '[^a-z\\s]', ' ')

    # shrink down to just one white space
    temp <- str_replace_all(temp,'[\\s]+', ' ')

    # split it
    temp <- str_split(temp, ' ')[[1]]
    
    # remove empty characters
    temp <- temp[!(temp %in% '')]

    return(temp)

}
```

```r
cleaning <- function(document) {

    # remove stop words
    stop_words <- unique(str_split(paste0(stopwords(), collapse = ' '), '[\\s|\']')[[1]])
    temp <- document[!(document %in% stop_words)]

    # remove single letters
    temp <- temp[!(temp %in% letters)]

    return(temp)

}
```

```r
pre_processing <- function(document, CLEAN = TRUE, STEM = FAlSE, LEM = FALSE) {

    # transform string in tokens
    document <- tokenize(document)

    # clean tokens
    if (CLEAN) { document <- cleaning(document) }

    # stem tokens
    if (STEM) { document <- stem_words(document) }

    # lemmatize tokens
    if (LEM) { document <- lemmatize_words(document) }


    return(document)
}


v_pre_processing <- Vectorize(pre_processing, vectorize.args = c('document'))
```

```r
make_vocabulary <- function(train) {

    # join every list of tokens into a unique list
    res <- do.call(c, train$Tokens)

    # create the vocabulary of words of the training set
    return(sort(unique(res)))

}
```

```r
get_mutual_information <- function(train, vocabulary) {

    # mutual information
    mutual_information <- list()
    
    # function to count number of documents containing a given token
    count_documents <- function(token, documents) {

        counts <- sum(sapply(documents, function(x) token %in% x))
        return(counts)

    }
                
    v_count_documents <- Vectorize(count_documents, vectorize.args = c('token'))

    for (label in label_list) {

        # number of documents that contain token t and have label l
        N_11 <- unname(v_count_documents(vocabulary, documents = train$Tokens[train$Labels == label]))
        # number of documents that do not contain token t and have label l
        N_01 <- nrow(train[train$Labels == label,]) - N_11

        # number of documents that contain token t and do not have label l
        N_10 <- unname(v_count_documents(vocabulary, documents = train$Tokens[train$Labels != label]))
        # number of documents that do not contain token t and do not have label l
        N_00 <- nrow(train[train$Labels != label,]) - N_10

        # total number of documents
        N <- N_11 + N_01 + N_10 + N_00

        # log losses
        log_loss_11 <- ifelse(N_11 > 0, log2(N*N_11/((N_10 + N_11)*(N_01 + N_11))), 0)
        log_loss_01 <- ifelse(N_01 > 0, log2(N*N_01/((N_00 + N_01)*(N_01 + N_11))), 0)
        log_loss_10 <- ifelse(N_10 > 0, log2(N*N_10/((N_10 + N_11)*(N_00 + N_10))), 0)
        log_loss_00 <- ifelse(N_00 > 0, log2(N*N_00/((N_00 + N_01)*(N_00 + N_10))), 0)
        
        # mutual information
        MI <- (N_11/N)*log_loss_11 +
            (N_01/N)*log_loss_01 + 
            (N_10/N)*log_loss_10 + 
            (N_00/N)*log_loss_00

        # add names
        names(MI) <- vocabulary

        # add mutual information to the list
        mutual_information[[paste0('Label_', label)]] <- MI
 
    }

    return(mutual_information)
         
}
```

```r
features_selection <- function(mutual_information, k) {

    # selected features
    selected_features <- list()

    for (label in label_list) {

        # sort
        MI <- sort(mutual_information[[paste0('Label_', label)]], decreasing = T)

        # select k [%] features
        MI <- MI[1:floor(k*length(MI))]

        # add selected features to the list
        selected_features[[paste0('Label_', label)]] <- MI

    }

    # join every named vector into a unique list
    res <- do.call(c, unname(selected_features))

    # create the new vocabulary of tokens of the training set after feature selection
    vocabulary <- sort(unique(names(res)))

    return(list(selected_features = selected_features,
                vocabulary = vocabulary))

}
```

```r
plot_MI <- function(mutual_information, k) {

    # plot ordered mutual information for each class
    options(repr.plot.width = 15, repr.plot.height = 10)
    par(mfrow = c(ceiling(length(label_list)/3), 3))

    for (label in label_list) {

        MI <- mutual_information[[paste0('Label_', label)]]
        MI[is.nan(MI)] <- 0
        MI <- sort(MI, decreasing = T)
        plot(1:length(MI), MI, ylim = c(0, 0.015), type = 'l', lwd = 2, col = 'navyblue',
            main = paste0('Label_', label), xlab = 'word index', ylab = 'MI',
            cex.lab = 1.5, cex.axis = 1.5, cex.main = 2)
        abline(v = floor(k*length(MI)), lwd = 2, lty = 2, col = 'red')       
    }
}
```

```r
learning_model <- function(train, vocabulary) {

    # create dataframe of conditional probabilities
    conditional_probs <- data.frame(Token = vocabulary)
    priors <- NULL
    tot_tokens_per_label <- NULL

    # number of training data
    N <- nrow(train)

    # function to compute unormalized conditional probability with Laplace smoothing
    conditional_prob <- function(token, tokens_list){

        n_token <- sum(tokens_list == token) + 1
        return(n_token)

    }
    
    v_conditional_prob <- Vectorize(conditional_prob, vectorize.args = c('token'))

    for (label in label_list) {
        
        # create list of tokens grouped by label
        tokens_list <- do.call(c, train$Tokens[train$Labels == label])

        # count total number of token in label l
        tot_tokens <- length(tokens_list)

        # save number of tokens per label (will be used in the test set)
        tot_tokens_per_label  <- c(tot_tokens_per_label, tot_tokens + length(vocabulary))

        # calculate the conditional probability function for every token of the vocabulary
        conditional_probs[[paste0('Label_', label)]] <- v_conditional_prob(conditional_probs$Token, tokens_list=tokens_list)

        # normalize the conditional probabilities
        normalization <- (tot_tokens+ length(vocabulary))
        conditional_probs[[paste0('Label_', label)]] <- conditional_probs[[paste0('Label_', label)]]/normalization

        # calculate prior for the labels
        N_l <- sum(train$Labels == label)
        priors <- c(priors, N_l/N)
    }

    return(list(conditional_probs = conditional_probs,
                priors = priors,
                tot_tokens_per_label = tot_tokens_per_label))

}
```

```r
predict <- function(document, vocabulary,
                    conditional_probs, priors, tot_tokens_per_label,
                    CLEAN = TRUE, STEM = FALSE, LEM = FALSE){

    # pre-process document
    tokens_in_document <- pre_processing(document, CLEAN = CLEAN, STEM = STEM, LEM = LEM)

    # select conditional probabilities of token in document
    conditional_probs <- conditional_probs[vocabulary %in% tokens_in_document,]

    # calculate log-likelihood of the document for each label
    log_likelihood <- colSums(log(conditional_probs[, 2:(1+length(label_list))]))

    # calculate log-posterior for each label 
    log_posterior <- log_likelihood + log(priors) 

    # count the number of tokens not in the vocabulary
    n_unseen_tokens <- length(setdiff(tokens_in_document, vocabulary))

    # add the log-posterior for the tokens not in the vocabulary
    log_posterior <- log_posterior - n_unseen_tokens*log(tot_tokens_per_label)

    predicted_label <- unname(which.max(log_posterior) - 1)

    return(predicted_label)

}

v_predict <- Vectorize(predict, vectorize.args = c('document'))
```

```r
get_confusion_matrix <- function(true_labels, predicted_labels) {

    predicted_labels <- unname(predicted_labels)

    # confusion matrix
    cm <- confusionMatrix(as.factor(predicted_labels), as.factor(true_labels), dnn = c("Prediction", "Reference"))
    return(cm)

}
```

```r
get_micro_f1_measure <- function(true_labels, predicted_labels) {

    label_count = aggregate(true_labels, by=list(true_labels), FUN=length)
    names(label_count) <- c('Labels', 'Count')

    # total true positive
    tot_TP <- 0
    # total false positive
    tot_FP <- 0
    # total false negative
    tot_FN <- 0

    for (label in label_count$Labels) {

        tot_TP <- tot_TP + sum(predicted_labels[true_labels == label] == label)
        tot_FP <- tot_FP + sum(predicted_labels[true_labels != label] == label)
        tot_FN <- tot_FN + sum(predicted_labels[true_labels == label] != label)

    }

    micro_precision = tot_TP/(tot_TP + tot_FP)
    micro_recall = tot_TP/(tot_TP + tot_FN)

    return(2*(micro_precision*micro_recall)/(micro_precision + micro_recall))
    }
```

```r
get_accuracy <- function(true_labels, predicted_labels) {
    return(sum(true_labels == predicted_labels)/length(true_labels))
}
```

```r
evaluate <- function(true_labels, predicted_labels) {

    return(list(confusion_matrix = get_confusion_matrix(true_labels, predicted_labels),
                micro_f1_measure = get_micro_f1_measure(true_labels, predicted_labels),
                accuracy = get_accuracy(true_labels, predicted_labels)))
                
}
```

```r
# plot confusion matrix
plot_confusion_matrix <- function(cm) {
    # divide each column by its sum
    plt <- as.data.frame(round(sweep(cm$table,2,colSums(cm$table),FUN="/"), 2))
    plt$Prediction <- factor(plt$Prediction, levels=rev(levels(plt$Prediction)))

    ggplot(plt, aes(Prediction,Reference, fill= Freq)) +
            geom_tile() + geom_text(aes(label=Freq), size=8) +
            scale_fill_gradient(low="white", high="#009194") +
            labs(x = "Reference",y = "Prediction") +
            scale_x_discrete(labels=c("Label_1","Label_2","Label_3","Label_4","Label_5","Label_6")) +
            scale_y_discrete(labels=c("Label_6","Label_5","Label_4","Label_3","Label_2","Label_1")) +
            theme(axis.text.x = element_text(size = 15),
                    axis.text.y = element_text(size = 15),  
                    axis.title.x = element_text(size = 15),
                    axis.title.y = element_text(size = 15)) +
                    theme(legend.position="none")

}
```

# Naive Bayes classifier for Fake News recognition

The goal of the project is to implement a Multinomial Naive Bayes classifier in R and test its performances in the classification of social media posts.

### Dataset

We analyze a [dataset](https://www.kaggle.com/datasets/anmolkumar/fake-news-content-detection?select=train.csv) available on Kaggle that consists of more than 10000 text documents representing content from social media and labeled as follows:

0 -> Barely-True \
1 -> False \
2 -> Half-True \
3 -> Mostly-True \
4 -> Not-Known \
5 -> True

```r
# set seed
seed <- 54321
set.seed(seed)

# import data
data <- read.csv('data/train.csv')

# select a subset of the data
#data <- data[sample(nrow(data), 1000), ]

# merge close labels
#data$Labels <- sapply(data$Labels, function(x) switch(x+1, 0, 0, 1, 2, 1, 2))

# get labels
label_list <- sort(unique(data$Labels))

# print some info
cat('number of rows:', nrow(data), '\n')
cat('number of columns:', ncol(data), '\n')
head(data, 5)
```

### Training, validation and test sets

The dataset is divided into training set, validation set and testing set.
The training set is used to learn the model from the data and the validation set provides unbiased evaluation while tuning the hyperparameters of the model. Finally, the test set is used to evaluate the performance of the best model.

```r
# split training, validation and test set
index <- sample(seq(1, 3), size = nrow(data), replace = TRUE, prob = c(.7, .15, .15))

train <- data[index == 1,]
val <- data[index == 2,]
test <- data[index == 3,]
```

```r
# plot the percentage frequency of each label for each set
plot_label_freq(list(training = train, validation = val, test = test))
```

### Tokenization

Each document in our dataset is a string of characters that must be preprocessed before being fed into the algorithm. In particular we convert the uppercase letters of a string into lowercase letters and then chop it up into chunks, called tokens, throwing away anything that is not a letter. This represents the first step in the preprocessing stage, which, however, is not without its problems. For example, chopping on apostrophe for possession and contractions could lead to incorrect tokens. These types of errors will be handled later. Here is an example of tokenization:

```r
document <- train$Text[20]
document_step1 <- tokenize(document)

cat('document:', '\n')
cat(document, '\n')

cat('\ntokenized document:', '\n')
print(document_step1)
```

### Stop words

The second step is to clean up any collection of tokens from stop words. The stop words are some extremely common words that seem to be of little value in helping to properly classify documents.  We also remove the single letters produced during the tokenization. Here is an example:

```r
document_step2 <- cleaning(document_step1)

cat('tokenized document:', '\n')
print(document_step1)

cat('\ntokenized document without stop words:', '\n')
print(document_step2)
```

### Lemmmatization

The last step in the preprocessing phase is what is known as token normalization, which consists of creating equivalence classes so that similar tokens are mapped to the same class. In particular, we use the lemmatization method, which performs a complete morphological analysis to accurately identify the lemma of each word, such as removing inflectional endings. ### GRID SEARCH

```r
# grid of values of k
k_grid <- seq(0.02, 1, 0.02)

# combination of filters
opt_grid <- list(c(F,F,F), c(T,F,F), c(T,T,F), c(T,F,T))

# accuracy grid
accuracy_grid <- matrix(0, nrow=length(opt_grid), ncol=length(k_grid))

if (FALSE) {
    for (o in 1:length(opt_grid)) {    
    
        # pre-process documents in training set
        train$Tokens <- v_pre_processing(train$Text,
                        CLEAN = opt_grid[[o]][1], STEM = opt_grid[[o]][2], LEM = opt_grid[[o]][3])

        # make vocabulary
        vocabulary <- make_vocabulary(train)
        
        # mutual information
        mutual_information <- get_mutual_information(train, vocabulary)
        
        
        
        for (i in 1:length(k_grid)) {
            
            
            # features selection
            k <- k_grid[i]
            res <- features_selection(mutual_information, k)

            selected_features <- res$selected_features
            vocabulary <- res$vocabulary
            
            
            # learning model
            res <- learning_model(train, vocabulary)

            conditional_probs <- res$conditional_probs
            priors <- res$priors
            tot_tokens_per_label <- res$tot_tokens_per_label
            
            # predict
            true_labels <- test$Labels
            predicted_labels <- sapply(test$Text, predict, 
                                vocabulary = vocabulary,
                                conditional_probs = conditional_probs, priors = priors,
                                tot_tokens_per_label = tot_tokens_per_label,
                                CLEAN = opt_grid[[o]][1], STEM = opt_grid[[o]][2], LEM = opt_grid[[o]][3],
                                USE.NAMES = F)
            
            # evaluate
            res <- evaluate(true_labels, predicted_labels)
            accuracy <- res$accuracy
            accuracy_grid[o, i] <- accuracy
            
        }
    }
}
```

### Feature selection: mutual information

The last stage before delving into the algorithm is represented by feature selection. Feature selection is the process of selecting a subset of the tokens in the vocabulary extracted from the training set and using only this subset as features in the classification of new documents. It is used to reduce the high-dimensional document space and eliminate noise features. This prevents the algorithm from overfitting and provides better generalization in the test set. In particular we compute a utility measure as the expected mutual information (MI) of each token and each label. MI measures how much information presence/absence of a term contributes to making the correct classification decision on a selected label. The formula is the following one:


$$

\text{MI}(X,Y) = \sum_{x \in X} \sum_{y \in Y} P_{X,Y}(x,y) \cdot \log_2\bigg(\frac{P_{X,Y}(x,y)}{P_{X}(x) \cdot P_{Y}(y)}\bigg)

$$

where $X$ and $Y$ are discrete variables, $P_{X,Y}(x,y)$ is the joint distribution and $P_{X}(x)$ and $P_{Y}(y)$ are the marginal distributions .

```r
# mutual information
mutual_information <- get_mutual_information(train, vocabulary)
```

```r
# evaluate
res <- evaluate(true_labels, predicted_labels)

f1_measure <- res$micro_f1_measure
accuracy <- res$accuracy
confusion_matrix <- res$confusion_matrix

cat('f1 measure:', f1_measure, '\n')
cat('accuracy:', accuracy, '\n')
cat('confusion matrix', '\n')
plot_confusion_matrix(confusion_matrix)
```
